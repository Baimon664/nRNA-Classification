```python
import numpy as np
names=open("npy_2/name.txt",'r')
m=[]
```

```python
q=3
def QMer(seq):
    data = []
    for i in range(len(seq)-(q-1)):
        data.append(list(seq[i:i+q]))
    return data

def genname(i):
  return "0"*(2-len(str(i)))+str(i)
```

```python
for i in names.readlines():
  i=i.strip()
  a=np.load("npy_2/"+i)
  for z in range(1,21):
    k=""
    p = a[z*-1]
    for j in range(len(p)):
      if(p[j]=="A" or p[j]=="a"):
        k+='1'
      elif(p[j]=="T" or p[j]=="t" or p[j]=="U" or p[j]=="u"):
        k+='2'
      elif(p[j]=="C" or p[j]=="c"):
        k+='3'
      elif(p[j]=="G" or p[j]=="g"):
        k+='4'
      else:
        k+='5'
    k+="0"*(200-len(k))
    data = QMer(k)
    m.append(data)
    # na = "3mer/"+i[:-4]+"_"+genname(z)+".npy"
    # np.save(na,t)
    # b=np.append(b,t)
  print(i,"finish")
```
